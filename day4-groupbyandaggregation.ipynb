```python
import pandas as pd

df = pd.DataFrame({
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "City": ["Delhi", "Mumbai", "Delhi", "Kolkata", "Mumbai", "Delhi"],
    "Salary": [50000, 40000, 60000, 70000, 45000, 55000],
    "Experience": [2, 5, 3, 7, 4, 6]
})

print(df)
```

```python
## groupby
df.groupby("City")["Salary"].mean()
```

```python
## count
df.groupby("Department").size()
```

```python
df.groupby("Department")["Employee"].count()  ## count ignores NaN
```

```python
## sum
df.groupby("Department")["Salary"].sum()
```

```python
## maximum
df.groupby("Department")["Salary"].max()
```

```python
## minimum
df.groupby("Department")["Salary"].min()
```

```python
## aggregation
df.groupby("Department")["Salary"].agg(["min","max","mean","median","sum","count"])
```

```python
## multiple column
df.groupby(
    ["Department", "City"]
)["Salary"].mean()
```

```python
## named aggregation
df.groupby("Department").agg(
    averagesalary = ("Salary", "mean"),
    maxsalary = ("Salary", "max"),
    minsalary = ("Salary", "min")
)
```

```python
## groupby multiple metric
data = df.groupby("Department").agg({
    "Salary" : ["mean", "max", "min"],
    "Experience" : ["mean", "max"]
})
```

```python
df.columns = [
    "".join(col)
    for col in data.columns
]
df
```

```python
## Transform
df["DeptAvgSalary"] = (
    df.groupby("Department")["Salary"]
      .transform("mean")
)
df
```

```python
df.groupby("Department").filter(
    lambda x: len(x) > 2
)
```

```python
pd.pivot_table(
    df,
    values="Salary",
    index="Department",
    aggfunc="mean"
)
```
