```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Alice", None],
    "Age": [20, 25, np.nan, 20, 28],
    "City": ["Delhi", "Mumbai", "Delhi", " ", " Kolkata "],
    "Salary": [30000, None, 45000, 30000, 60000]
})

print(df)
```

```python
## missing value
df.isnull().sum()
```

```python
## drop missing value
df.dropna()
```

```python
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Alice", None],
    "Age": [20, 25, np.nan, np.nan, 28],
    "City": ["Delhi", "Mumbai", "Delhi", " ", " Kolkata "],
    "Salary": [30000, None, 45000, 30000, 60000]
})

print(df)
```

```python
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Alice", None],
    "Age": [20, 25, np.nan, np.nan, 28],
    "City": ["Delhi", "Mumbai", "Delhi", " ", " Kolkata "],
    "Salary": [30000, None, 45000, 30000, 60000]
})

print(df)
```

```python
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Alice", None],
    "Age": [20, 25, np.nan, 20, 28],
    "City": ["Delhi", "Mumbai", "Delhi", "Delhi", " Kolkata "],
    "Salary": [30000, None, 45000, 30000, 60000]
})

print(df)
## Detect Duplicate Rows
df.duplicated()
```

```python
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Alice", None],
    "Age": [20, 25, np.nan, 20, 28],
    "City": ["Delhi", "Mumbai", "Delhi", "    Delhi", " Kolkata     "],
    "Salary": [30000, None, 45000, 30000, 60000]
})
print(df)
# Remove Extra Spaces
df["City"].str.strip()
df
```

```python
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Alice", None],
    "Age": [20, 25, np.nan, 20, 28],
    "City": ["DelhI", "MUMbai", "DeLhi", "    Delhi", " KOlKata     "],
    "Salary": [30000, None, 45000, 30000, 60000]
})
## Change text case
df["City"].str.upper().str.strip()
```

```python
## for lower case 
df["City"].str.lower().str.strip()
```

```python
df["City"].str.title()
```

```python
## replace values
df["City"] = df["City"].str.title().str.strip()
print(df)
df["City"] = df["City"].replace("Delhi","NewDelhi")
df
```

```python
## Data Type Conversion
df.dtypes
```

```python
df["Salary"] = df["Salary"].astype("f16")
df
```

```python
df["Age"] = pd.to_numeric(
    df["Age"],
    errors="coerce"
)
df
```

```python
## Unique values
df["City"].unique()
```

```python
df["City"].nunique()  ## count unique
```

```python
## Value Counts
df["City"].value_counts()
```

```python
## Rename column
df.rename(columns = {"City":"Town"}, inplace=True)
df
```

```python
## Create new column
df["YearlySalary"] = df["Salary"]*12
```
