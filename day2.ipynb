```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [20, 25, 22, 30, 28],
    "City": ["Delhi", "Mumbai", "Delhi", "Kolkata", "Mumbai"],
    "Salary": [30000, 50000, 45000, 70000, 60000]
})

print(df)
```

```python
## selecting column(single)
df["Age"]  ## series
```

```python
# selecting column (multiple)
df[["Name", "Salary"]] ## dataframe
```

```python
## selecting rows using loc and iloc

### iloc

df.iloc[0]  ## 1st row
```

```python
df.iloc[1]  ## 2nd row
```

```python
df.iloc[:3] ## first 3 rows
```

```python
df.iloc[1:4]  ## 2nd to 4th row
```

```python
df.iloc[[0, 2, 4]]  ## specific row
```

```python
df.iloc[0:3, 2:3]  ## [row, col]
```

```python
## loc = label
df.loc[0]  ## same as iloc
```

```python
df.loc[1]  ## same as iloc
```

```python
df.loc[:3] ## 4th row included
```

```python
df.loc[1:4] ## 2nd to 5th row
```

```python
df.loc[[0, 2, 4]]  ## same as iloc
```

```python
## df.loc[row, column] wont work unless column names are specified 
```

```python
## filtering rows
df[df["Salary"]>50000]
```

```python
df[df["City"] == "Delhi"]
```

```python
## multiple condition (and = & and or = |)
df[(df["Salary"]>50000) & (df["City"] == "Mumbai")]
```

```python
df[(df["City"] == "Delhi") | (df["City"] == "Mumbai")]
```

```python
## isin()  ## clean, simple and fast for big datasets
df[df["City"].isin(["Delhi", "Mumbai"])]
```

```python
## between
df[df["Age"].between(20,25)]
```

```python
## query()
df.query("Age<=25 and Age>=20")
```

```python
## sorting  ## default ascending
df.sort_values(["Salary"])
```

```python
## descencding
df.sort_values(["Salary"], ascending=False)
```

```python
## sort by 2 columns
df.sort_values(["City", "Age"])
```

```python
## rename column name
df.rename(columns={"Salary": "Monthly_salary"})
```
